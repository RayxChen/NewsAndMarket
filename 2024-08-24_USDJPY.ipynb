```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
usd_jpy = yf.download("JPY=X", start="2024-01-01", end="2024-08-24")
nikkei = yf.download("^N225", start="2024-01-01", end="2024-08-24")
```

```python
# Plotting the data with improved aesthetics
fig, ax1 = plt.subplots(figsize=(18, 10))

# Plot USD/JPY on the first y-axis
ax1.plot(usd_jpy.index, usd_jpy['Close'], label='USD/JPY Exchange Rate', color='royalblue', linewidth=2)
ax1.set_xlabel('Date', fontsize=14, fontweight='bold')
ax1.set_ylabel('USD/JPY Exchange Rate', color='royalblue', fontsize=14, fontweight='bold')
ax1.tick_params(axis='y', labelcolor='royalblue')
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)

# Create a second y-axis for the Nikkei 225 index
ax2 = ax1.twinx()
ax2.plot(nikkei.index, nikkei['Close'], label='Nikkei 225 Index', color='darkred', linewidth=2)
ax2.set_ylabel('Nikkei 225 Index', color='darkred', fontsize=14, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='darkred')
ax2.tick_params(axis='y', labelsize=12)

# Adding important dates with improved formatting
important_dates = {
    '2024-04-23': 'Japan issued strongest FX warning',
    '2024-05-04': 'Yellen refused to comment on yen support',
    '2024-06-20': 'US Treasury added Japan to FX watchlist',
    '2024-07-31': 'Kanda resignation takes effect'
}

for date, label in important_dates.items():
    ax1.axvline(pd.to_datetime(date), color='green', linestyle='--', lw=2)
    ax1.text(pd.to_datetime(date), ax1.get_ylim()[1] * 0.95, label, color='red',alpha=0.75, rotation=45,
             verticalalignment='top',horizontalalignment='center', fontsize=12, fontweight='bold')

# Adding title and grid
plt.title('USD/JPY Exchange Rate and Nikkei 225 Index (2024)', fontsize=16, fontweight='bold')
ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot with a tight layout
fig.tight_layout()
plt.show()
```
